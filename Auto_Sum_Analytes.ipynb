```python
import pandas as pd
from openpyxl import load_workbook
```

```python
# Import the Excel worksheet
path = "C:\\Users\\MJolliffe\\Documents\\Python_Projects\\Analyte_Sums\\RM16.5_URSC_DataWithSums_022318_2.xlsx"
xl = pd.ExcelFile(path)
df = xl.parse("Data")

df.head()
```

```python
# Re-index by sys_sample_code (location) and analyt group
df_by_proj = df.set_index(['Project','sys_sample_code','AnalyteGroup','chemical_name'])
df_by_proj.sort_index(inplace=True)
df_by_proj.head()
#df_by_proj.xs('RM16.5-COMP-A', level='sys_sample_code')
```

```python
# See what a python dict looks like
#analytes_dict = df_by_proj.T.to_dict()
truncated = pd.DataFrame(df_by_proj[['validation_qualifiers','GSI_Result']])

flat_dictionary = truncated.to_dict()
#outerkey1 = truncated.index.get_level_values('Project')
flat_dictionary
```

```python
# build nested dictionary in the form {Project: {Site: {paramGroup: {param: [value, flag]}}}}
d = {} 
for (project, site, grouping, param), value in flat_dictionary['GSI_Result'].items():
    d.setdefault(project, {}).setdefault(site, {}).setdefault(grouping, {})[param] = [value]

for (project, site, grouping, param), flag in flat_dictionary['validation_qualifiers'].items():
    d[project][site][grouping][param].append(flag)

d
    
```

```python
# Sum by parameter group, by site, by project 
sum_df = pd.DataFrame(columns=['Project','sys_sample_code','AnalyteGroup','GSI_Sum'])
for project in d:
    for site in d[project]:
        for paramgroup in d[project][site]:
            # Sum the individual parameter values, halving if flag is a 'U'
            sum_df = sum_df.append({'Project': project, 'sys_sample_code': site, 'AnalyteGroup': "Total {}".format(paramgroup), 'GSI_Sum': sum([1/2 * val[0] if val[1] == 'U' else val[0] for key, val in d[project][site][paramgroup].items()])}, ignore_index = True)
            print("Total {} at {}, {}:  {}".format(paramgroup, project, site, sum([1/2 * val[0] if val[1] == 'U' else val[0] for key, val in d[project][site][paramgroup].items()]))) 
            
sum_df.head(10)            
```
