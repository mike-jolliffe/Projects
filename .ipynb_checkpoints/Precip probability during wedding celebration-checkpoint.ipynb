```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Precip_pdx.csv')
data.head(30)
```

```python
days = []
total = 0
no_rain = 0
rain = []
for value in data["22"]:
    if value[0].isdigit():
        total += 1
        days.append(value)
        if value[0] == "0":
            no_rain += 1
        else:
            rain.append(int(value))
avg_rain = np.mean(rain)
            
print("Total period of record is {} years".format(total))
print ("{}% chance of not raining on 6/22".format(((no_rain/total)*100)))
print ("Average rainfall when it did rain was {} mm".format(avg_rain))
```
